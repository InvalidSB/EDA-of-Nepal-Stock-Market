```python
from matplotlib.pyplot import figure

figure(figsize=(12, 8), dpi=80)
plt.title("Nepal Whole stock price",fontsize=44,color="w")
df['nepse open'].plot(label="opening price")
df['nepse high'].plot(label="high price",color="r")
df['nepse Low'].plot(label="Low price")
df['nepse close'].plot(label="closing price",color='k')
plt.legend()
plt.show()
```

```python
#variation aayeko xa vanne ,opening and closing price ma different dekhesi tha hunax
paxikodf=df.iloc[5297:]
#paxikodf
from matplotlib.pyplot import figure

figure(figsize=(12, 8), dpi=80)
plt.title("Nepal stock in 2020-9 to 2021-7",fontsize=44,color="w")
paxikodf['nepse open'].plot(label="opening price")
paxikodf['nepse high'].plot(label="high price",color="r")
paxikodf['nepse Low'].plot(label="Low price")
paxikodf['nepse close'].plot(label="closing price",color='b')
plt.legend()
plt.show()
```

```python
#daily percentage change
df['returns']=(df['nepse close']/df['nepse close'].shift(1))-1
```

```python
df['returns']
```

```python
df['returns'].hist(bins=200)
# good lagxa if normal distribution jasto vaye
```

```python
df['returns'].plot(kind="kde")
plt.show()
```

```python
#cumilative returns
df["cumilativeReturns"]=(1+ df['returns']).cumprod()
df.tail(5)
```

```python
df["cumilativeReturns"].plot(figsize=(15,10))

plt.show()
# this gives =>if we invest Rs 1  at begining then it will returns Rs 17.65 today
```
